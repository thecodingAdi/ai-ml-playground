```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

<!-- Cleaned: Clone Repository section -->

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
```

<!-- Cleaned: Add Changes and Push section -->

```python
model_knn.fit(X_train, y_train)
```

<!-- Cleaned: Save and Push specific files section -->

```python
model_knn.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(C=30, gamma='auto', kernel= 'rbf')
```

<!-- Cleaned: Create and Push a New Working File section -->

```python
model_svm.fit(X_train, y_train)
```

<!-- Cleaned: Push to GitHub instructions -->

```python
model_svm.score(X_test, y_test)
```

```python
 # Using Grid Search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svm),{
    'C' : [1,10,20,30],
    'kernel' : ['rbf', 'linear']
}, cv=5, return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result
```

```python
result[['param_C','param_kernel','mean_test_score']]
```

```python
classifier2 = GridSearchCV((model_knn),{
    'n_neighbors' : [1,5,10],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'metric' : ['minkowski', 'euclidean', 'manhattan'],
    'weights' : ['uniform', 'distance']
}, cv=5, return_train_score=False)
```

```python
classifier2.fit(X,y)
```

```python
classifier2.cv_results_
```

```python
result2 = pd.DataFrame(classifier2.cv_results_)
result2
```

```python
result2[['param_n_neighbors','param_algorithm','param_metric','param_weights','mean_test_score']]
```
