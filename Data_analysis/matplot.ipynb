##### Using distribution plots to visualize the distribution of quantitative data.

 some plots that allow us to visualize the distribution of a data set. These plots are:

- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = sns.load_dataset("tips")
df
```

```python
df['size'].unique()

plt.subplot(1,2,1)  # 1-row, 2-col, 1-left side position

sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```
