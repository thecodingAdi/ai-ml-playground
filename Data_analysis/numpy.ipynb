```python
import numpy as np

x = np.array([[1, 2, 3],
              [4, 5, 6]], dtype=np.float32)

print(x.shape)   # (2 rows, 3 col)
print(x.dtype)   # float32
print(x.ndim)    # 2
print(x.sum(axis=0)) #[5. 7. 9.]
print(x.sum(axis=1)) # [ 6. 15.]

mean = x.mean(axis=0)
print(mean, mean.shape)
```

### Normalization on array

```python
import numpy as np

X = np.array([[70,80,90],
         [60,75,85],
         [90,95,100]])

mean = X.mean(axis=0) # calculating mean
std = X.std(axis=0) # calculating Standard deviation

X_norm = (X - mean)/std # using Normalization formula (Normal Distribution)

# printing result
print("Original Data:\n", X,"\n")
print("Mean:", mean,"\n")
print("Std:", std,"\n")
print("Normalized Data:\n", X_norm,"\n")

#  Verify normalization
print("Mean after normalization:", X_norm.mean(axis=0))
print("Std after normalization:", X_norm.std(axis=0))
```
