## Pandas Series 
##### A seires is a 1-D labeled array capable of holding any data type. The axis lables are collectively called the index

### Creating Series
#### there are numbers of ways to create a pandas series


```python
import numpy as np
import pandas as pd
```

```python
labels =  ['a','b','c']
arr = [10,20,30]
d = {1:10,2:20,3:30}
```

```python
pd.Series(labels)
```

```python
pd.Series(arr,index=labels)
```

```python
pd.Series(d,index = labels)
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "London", "Paris"]
}

pd.DataFrame(data)
```
