### Abstraction
---

```python
from abc import ABC, abstractmethod
class Greet(ABC):
    @abstractmethod
    def say_hello(self):
        pass
class English(Greet):
    def say_hello(self):
        return "Hello There!"

g1 = English()
print (g1.say_hello())
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def draw(self):
        pass
    
    @abstractmethod
    def area(self):
        pass
```

```python
class square(Shape):
    def draw(self):
        return "Drawing Square with area = "
    def area(self,l):
        return (l*l)

class circle(Shape):
    def draw(self):
        return "Drawing Circle with area = "
    def area(self,r):
        return (3.14*(r**2))

class Triangle(Shape):
    def draw(self):
        return "Drawing Triangle"

class Star(Shape):
    def draw(self):
        return "Drawing Star"

sq = square()
ci = circle()
print(sq.draw(),sq.area(5))
print(ci.draw(),ci.area(7))
```

### Encapsulation
---

```python
from abc import ABC, abstractmethod

class Bank(ABC):
    @abstractmethod
    def account(self, acc_num):
        pass

    @abstractmethod
    def balance(self, bal):
        pass

    @abstractmethod
    def transaction(self):
        pass
```

```python
class Deposit(Bank):
    def __init__(self):
        self.acc_num = None
        self.bal = 0

    def account(self, acc_num):
        self.acc_num = acc_num

    def balance(self, bal):
        self.bal += bal

    def transaction(self):
        print("Transaction completed")


# object creation
dep = Deposit()

dep.account(123456789)
dep.balance(100000)

print("Account Number:", dep.acc_num)
print("Balance:", dep.bal)
```

### Encapsulation part 

```python
from abc import ABC, abstractmethod

class Bank(ABC):
    @abstractmethod
    def account(self, acc_num):
        pass

    @abstractmethod
    def deposit(self, amount):
        pass


class Deposit(Bank):
    def __init__(self):
        self.__acc_num = None   # private
        self.__bal = 0          # private

    def account(self, acc_num):
        self.__acc_num = acc_num

    def deposit(self, amount):
        if amount > 0:
            self.__bal += amount
        else:
            print("Invalid amount!")

    # Getter methods
    def get_balance(self):
        return self.__bal

    def get_account(self):
        return self.__acc_num


# usage
dep = Deposit()
dep.account(123456789)
dep.deposit(100000)

print("Account:", dep.get_account())
print("Balance:", dep.get_balance())
```
