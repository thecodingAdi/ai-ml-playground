# Z-Score Test Example
This notebook demonstrates how to calculate the Z-score and perform a one-sample Z-test in Python.

```python
import numpy as np
import scipy.stats as stats

# Scenario: 
# A school principal claims that the students in his school score an average of 70 out of 100 in exams. 
# The standard deviation of the population is known to be 15. 
# A random sample of 30 students has a mean score of 78.
# We will test if there is a significant difference using a Z-test (alpha = 0.05).

population_mean = 70.0
population_std = 15.0
sample_mean = 78.0
n = 30

# 1. Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
print(f"Z-score: {z_score:.4f}")

# 2. Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
print(f"P-value: {p_value:.4f}")

# 3. Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nDecision: Reject the null hypothesis.")
    print("Conclusion: The sample mean is significantly different from the population mean.")
else:
    print("\nDecision: Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant difference between the sample mean and the population mean.")
```
